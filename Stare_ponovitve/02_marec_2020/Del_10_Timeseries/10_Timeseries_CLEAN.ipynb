## Resampling

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head()
```

```python
opsd_weekly_mean.index
```

```python
start, end = '2017-01', '2017-06'
```

```python
fig, ax = plt.subplots()

ax.plot(opsd_daily.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       linewidth=0.5,
       label='Daily')

ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       markersize=8,
       label='Weekly Mean Resample')


ax.set_ylabel('Solar Production (GWh)')
ax.legend()
plt.show()
```

```python
opsd_monthly = opsd_daily[data_columns].resample('M').sum(min_count=28)
```

```python
opsd_monthly.head()
```

```python
fig, ax = plt.subplots()

ax.plot(opsd_monthly['Consumption'],
        color='black',
       label='Consumption')

opsd_monthly[['Wind', 'Solar']].plot.area(ax=ax, linewidth=0)

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total (GWh)')
plt.show()
```

```python
opsd_annual = opsd_daily[data_columns].resample('A').sum(min_count=360)
```

```python
opsd_annual.head()
```

```python
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual.index.name = 'Year'
```

```python
opsd_annual.head()
```

```python
opsd_annual['Wind+Solar/Consumtion'] = opsd_annual['Wind+Solar']/opsd_annual['Consumption']
```

```python
opsd_annual.tail(3)
```

```python
ax = opsd_annual.loc[2012:, 'Wind+Solar/Consumtion'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation=0)
plt.show()
```

## Rolling windows

<img src="images/rollingwindow.png">

```python
opsd_7d = opsd_daily[data_columns].rolling(7, center=True).mean()
opsd_7d.head(12)
```

```python
start, end = '2017-01', '2017-06'


fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       linewidth=0.5,
       label='Daily')

ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
       marker='o',
       markersize=8,
       linestyle='-',
       label='Weekly Mean Resample')

ax.plot(opsd_7d.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       label='7-d Rolling Mean')

ax.set_ylabel('Solar Production (GWh)')
ax.legend()
plt.show()
```

## Trends

```python
opsd_365d = opsd_daily[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
fig, ax = plt.subplots()

ax.plot(opsd_daily['Consumption'],
       marker='.',
       markersize=2,
       color='0.6',
       linestyle='None',
       label='Daily')

ax.plot(opsd_7d['Consumption'],
       linewidth=2,
       label='7-d Rolling Mean')

ax.plot(opsd_365d['Consumption'],
       linewidth=3,
       color='0.2',
       label='Trend (365-d Rolling Mean)')

ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.show()
```

```python
fig, ax = plt.subplots()

for nm in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(opsd_365d[nm], label=nm)
    
    
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.set_ylim(0,400)
ax.legend()
ax.set_ylabel('Production (GWh)')
ax.set_title('Trends in Electricity Production (365-d Rolling Means)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.show()
```
